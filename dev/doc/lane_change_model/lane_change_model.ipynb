```python
import sympy as sy
from sympy.plotting import plot, plot_parametric
```

```python
sy_t = sy.symbols('t')
sy_c = (sy.cos(sy_t), sy.sin(sy_t))
sy_o = 0.5*sy_t
```

```python
plot_parametric(sy_c, (sy_t, 0,1), aspect_ratio=(1,1), ylim=(0,1), xlim=(0,1));
plot(sy_o, (sy_t, 0,1), aspect_ratio=(1,1), ylim=(0,1), xlim=(0,1));
```

```python
sy_c_d = tuple([sy.diff(x) for x in sy_c])
sy_c_n = (-sy_c_d[1], sy_c_d[0])
sy_cp = tuple([sy_c[i] + (sy_o * sy_c_n[i]) for i in range(2)])
print(sy_cp)
```

```python
p1 = plot_parametric(sy_cp, (sy_t, 0,1), aspect_ratio=(1,1), ylim=(0,1), xlim=(0,1), show=False)
p2 = plot_parametric(sy_c, (sy_t, 0,1), aspect_ratio=(1,1), ylim=(0,1), xlim=(0,1), show=False)
p1.append(p2[0])
p1.show()
```
